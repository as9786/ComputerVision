# Sobel filter

```python
import numpy as np
from PIL import Image

import scipy.signal as signal
```

```python
#  
img = Image.open('example.jpg')
img
```

```python
# Horizontal
sobel_h = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

# Vertical
sobel_v = np.array([[-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]])
```

```python
#  
img = img.convert('L')
# Image to ndarray
img = np.array(img)
img.shape
```

```python
# Apply Sobel filter
out_h = signal.convolve2d(img, sobel_h, boundary='symm', mode='same')
out_v = signal.convolve2d(img, sobel_v, boundary='symm', mode='same')

#  
out = np.sqrt(np.square(out_h) + np.square(out_v))

#  0-255  
out = out / out.max() * 255

# ndarray to Image
out_img = Image.fromarray(out.astype(np.uint8))
out_img
```
