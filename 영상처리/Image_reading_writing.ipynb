# Module

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

# 1. 사진 불러오기

```python
img = cv2.imread('/content/다운로드.jfif', cv2.IMREAD_COLOR)
```

### 첫 번째 인자
- 'test.jpg' : 읽고자 하는 사진

### 두 번째 인자
- cv2.IMREAD_COLOR : Load color image. 투명한 부분은 모두 무시
- cv2.IMREAD_GRAYSCALE : 흑백 영상
- cv2.IMREAD_UNCHANGED : Alpha-channel을 폼하하여 영상 불러오기
